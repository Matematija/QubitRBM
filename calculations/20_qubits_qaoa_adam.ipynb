```python
from QubitRBM.qaoa import QAOA
from QubitRBM.utils import hilbert_iter
from QubitRBM.optimize import qaoadam
```

```python
nq = 20
k = 3
```

```python
G = nx.random_regular_graph(k, nq)
```

```python
nx.draw_spring(G, node_color='gold', node_size=500, with_labels=True)
```

```python
qaoa = QAOA(graph=G, p=2)
```

```python
hilbert = np.array(list(hilbert_iter(nq)))
```

```python
%%time
init = [np.pi/8, np.pi/8, 0, 0]
par, his = qaoadam(qaoa, init=init, lr=1e-3, tol=1e-5, hilbert=hilbert)
```

```python
gamma_0, gamma_1, beta_0, beta_1 = par
```
